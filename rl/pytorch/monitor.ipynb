```python
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
sns.set()
```

```python
load_id = 9
seed = 2
env_id = 'StochasticTrack-v0'
rl_alg = 'SAC'
```

```python
fig = plt.figure(figsize=(6, 20))
ax1 = fig.add_subplot(511)
ax2 = fig.add_subplot(512)
ax3 = fig.add_subplot(513)
ax4 = fig.add_subplot(514)
ax5 = fig.add_subplot(515)
plt.ion()
fig.show()
fig.canvas.draw()

while True:
    history = np.load(f'out/{env_id}/{rl_alg}_id{load_id}/seed{seed}.npy').item()
    steps = history['training_steps']
    returns = history['eval_return']
    losses_dict = history['losses']
    
    ax1.clear()
    ax1.set_xlabel('steps')
    ax1.set_ylabel('return')
    ax1.set_title('evaluation')
    ax1.plot(steps, returns)
    
    ax2.clear()
    ax2.set_xlabel('steps')
    ax2.set_ylabel('value loss')
    ax2.plot(steps, losses_dict['value_loss'])
    
    ax3.clear()
    ax3.set_xlabel('steps')
    ax3.set_ylabel('policy loss')
    ax3.plot(steps, losses_dict['policy_loss'])
    
    ax4.clear()
    ax4.set_xlabel('steps')
    ax4.set_ylabel('q1 loss')
    ax4.plot(steps, losses_dict['q1_loss'])
    
    ax5.clear()
    ax5.set_xlabel('steps')
    ax5.set_ylabel('q2 loss')
    ax5.plot(steps, losses_dict['q2_loss'])
    
    fig.canvas.draw()
    time.sleep(30)
    
```
